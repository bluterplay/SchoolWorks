## 2 Números aleatorios

```python
from numpy.random import randint
```

```python
def numerosUnicos(n):
    np.random.seed(123)
    lista = [randint(0,n*50,1)]
    i= 1
    unico = False
    while (i<n):
        while(unico== False):
            x= randint(0,n*50,1)
            unico = True
            for j in lista:
                if x== j:
                    unico = False
                    break
        lista.append(x)
        i+= 1
        unico = False
    return lista
                    
                    
    
```

```python
ls_prueba= numerosUnicos(30)
ls_prueba
```
